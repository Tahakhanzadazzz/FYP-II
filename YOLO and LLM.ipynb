```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import requests

model = YOLO("/content/yolov8m_trained.pt")

def PredecitObj(image):

  results = model.predict(
      source=image,
      conf=0.25,
  )

  detected_objects_info = []
  for box in results[0].boxes:
      detected_objects_info.append({
          'class': results[0].names[int(box.cls)],
      })

  return detected_objects_info
```

```python
correct_predictions = 0
total_pairs = len(UpdatedData["pairs"])
true_positives = 0
false_positives = 0
false_negatives = 0

for key, pair in UpdatedData["pairs"].items():
    expected_response = 1 if pair["subtask1_label"] == "YES" else 0
    predicted_response = int(json.loads(pair["response"])['response'])

    if predicted_response == expected_response:
        correct_predictions += 1
    if predicted_response == 1 and expected_response == 1:
        true_positives += 1
    elif predicted_response == 1 and expected_response == 0:
        false_positives += 1
    elif predicted_response == 0 and expected_response == 1:
        false_negatives += 1

accuracy = (correct_predictions / total_pairs) * 100

precision = true_positives / (true_positives + false_positives) if (true_positives + false_positives) > 0 else 0
recall = true_positives / (true_positives + false_negatives) if (true_positives + false_negatives) > 0 else 0

f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

print(f"Accuracy: {accuracy:.2f}%")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1_score:.2f}")
```
